```python
enginelocation = pd.get_dummies(df_1['enginelocation'] , drop_first = 'True')
df_1 = pd.concat([df_1,enginelocation], axis = 1).drop(['enginelocation'],axis = 1)
df_1
```

```python
enginetype = pd.get_dummies(df_1['enginetype'] , drop_first = 'True')
df_1 = pd.concat([df_1,enginetype], axis = 1).drop(['enginetype'],axis = 1)
df_1
```

```python
cylindernumber = pd.get_dummies(df_1['cylindernumber'] , drop_first = 'True')
df_1 = pd.concat([df_1,cylindernumber], axis = 1).drop(['cylindernumber'],axis = 1)
df_1
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
